## Ejemplo Repositorio

Clonar una librer√≠a y utilizarla implica copiar el repositorio completo de la librer√≠a desde un servidor de control de versiones, como GitHub o GitLab, a la m√°quina local. Esto significa que se descargan todos los archivos y la historia del repositorio, incluyendo cambios, versiones anteriores, etc. Despu√©s de clonar la librer√≠a, el usuario puede editarla localmente y aplicar cualquier cambio que desee, y luego utilizarla en su proyecto.

```python
!git clone https://github.com/sloria/TextBlob
```

Al hacer esto podemos ver que en el sistema de archivos se ha creado una carpeta 'TextBlob' que contiene los mismos archivos que el repositorio:

[TextBlob](https://github.com/sloria/TextBlob)

```python
ls
```

Accedemos a ella y utilizamos una de las clases que vienen dentro de los archivos, en concreto la clase *TextBlob* del archivo *blob.py*

```python
cd TextBlob
```

```python
from textblob.blob import TextBlob

# Frase a analizar
es_text = "Me encanta escuchar m√∫sica"

# Codificamos con textblob y traducimos al ingl√©s
text = TextBlob(es_text).translate(from_lang='es', to='en')

# Analizamos la polaridad de las palabras
sentiment = text.sentiment.polarity

if sentiment > 0.0:
	print('üëç Positivo: ', sentiment)
else:
	print('üëé Negativo: ', sentiment)
```

## Ejemplo librer√≠a

Instalar una librer√≠a con pip install implica descargar e instalar solo la versi√≥n m√°s reciente de la librer√≠a desde un repositorio de paquetes de Python como PyPI (Python Package Index). La librer√≠a se instala en un directorio espec√≠fico del sistema y se hace disponible para su uso en cualquier proyecto que utilice Python. No se descargan los archivos de la librer√≠a en su totalidad y, por lo tanto, no se puede editar la librer√≠a localmente.

```python
!pip install textblob
```

```python
# Se importa de la librer√≠a el componente textblob
from textblob import TextBlob

# Frase a analizar
es_text = "Me encanta escuchar m√∫sica"

# Codificamos con textblob y traducimos al ingl√©s
text = TextBlob(es_text).translate(from_lang='es', to='en')

# Analizamos la polaridad de las palabras
sentiment = text.sentiment.polarity

if sentiment > 0.0:
	print('üëç Positivo: ', sentiment)
else:
	print('üëé Negativo: ', sentiment)
```
